```python
# Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels as sm
import os
import sys
from pathlib import Path

file_path = Path("../Data")
path_1 = file_path
```

```python
baseline_path = file_path / "2013-0533_data_baseline.dta"
baseline = pd.read_stata(baseline_path)
```

```python
# STEP 1: Define variable groups (same as your Stata locals)
hh_composition = ["hh_size", "adults", "children", "male_head", "head_age", "head_noeduc"]
credit_access = ["spandana", "othermfi", "bank", "informal", "anyloan"]
loan_amt = ["spandana_amt", "othermfi_amt", "bank_amt", "informal_amt", "anyloan_amt"]
self_emp_activ = ["total_biz", "female_biz", "female_biz_pct"]
businesses = ["bizrev", "bizexpense", "bizinvestment", "bizemployees", "hours_weekbiz"]

# STEP 2: Create new business outcome variables for all households
businesses_allHH = []
allvars = []

# Create new variables with "_allHH" suffix for all households
for var in businesses:
    new_var = f"{var}_allHH"
    businesses_allHH.append(new_var)
    baseline[new_var] = baseline[var]  # Copy original variable to new column
    baseline.loc[baseline["total_biz"] == 0, new_var] = 0  # Replace with 0 if no business

# Add the new _allHH variables to the allvars list
allvars.append(businesses_allHH)  # Append the new variables

consumption = ["total_exp_mo", "nondurable_exp_mo", "durables_exp_mo", "home_durable_index"]

# Combine all variable groups into one list
allvars = hh_composition + credit_access + loan_amt + self_emp_activ + businesses + businesses_allHH + consumption
```
